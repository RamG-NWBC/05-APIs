```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

circle1 = plt.Circle((0, 0), 5000, color='r', fill=False)
ax.add_artist(circle1)
ax.text(0, 0, "H", ha="center", va="center", fontsize=24)

ratio = 4000000/5000
circle2 = plt.Circle((5000*ratio, 0), 4000000, color='b', fill=False)
ax.add_artist(circle2)
ax.text(5000*ratio, 0, "A", ha="center", va="center", fontsize=24)

plt.xlim(-10000, 10000*ratio)
plt.ylim(-10000, 10000)
plt.show()
```
